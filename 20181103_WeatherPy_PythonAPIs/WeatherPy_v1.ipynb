```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import datetime
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# getting current date
date = datetime.datetime.today().strftime('%m-%d-%Y')
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
# list for holding cities based on random sample of lat, lon combos
cities = []
# checking for duplicates
inList = set()

# storing unique city, country in list of dicts
uniqueCities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1550)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1550)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    country = citipy.nearest_city(lat_lng[0], lat_lng[1]).country_code
    cities.append({'city': city, 'country': country})

# removing duplicates from the list of dictionaries 'cities' and creating a new list of dictionaries 'uniqueCities'
for dictx in cities:
    tuplex = tuple(dictx.items())
    if tuplex not in inList:
        inList.add(tuplex)
        uniqueCities.append(dictx)

# Print the city count to confirm sufficient count
print(len(cities))
print(len(uniqueCities))
print(uniqueCities)
```

## Perform API Calls

```python
# # TESTING ONE CITY AT A TIME -- TO BE COMMENTED OUT AT SOME POINT 
# url = 'https://api.openweathermap.org/data/2.5/weather?q=ballwin,us&units=imperial&appid=' + api_key
# response = requests.get(url).json()
# pprint(response)


# LOOPING FOR WHOLE SET OF CITIES IN UNIQUE CITIES
# setting up variables to hold list of responses for lat, temperature, humidity, cloudiness, windspeed
name = []
ctry = []
latitude = []
temperature = []
humidity = []
cloudiness = []
windspeed = []
counter = 0
url = 'https://api.openweathermap.org/data/2.5/weather?q='
separator = '-' * 30
# sending a request for each city in my list of dicts

print('Beginning Data Retrieval')
print(separator)
for city in uniqueCities:
    try:
        counter += 1
        # storing a search term
        query_url = f'{url}{city["city"]},{city["country"]}&units=imperial&appid={api_key}'
        # get the data 
        response = requests.get(query_url).json()
        #store name, country, lat, temp, humidity, cloudiness and windspeed in list variables
        name.append(response['name'])
        ctry.append(response['sys']['country'])
        latitude.append(response['coord']['lat'])
        temperature.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        windspeed.append(response['wind']['speed'])
        # print the processing of each city and country
        print(f'Processing Record {counter} | {city["city"]}, {city["country"]}')
        time.sleep(1)
    except KeyError:
        print("City not found. Skipping this record.")
        pass
print(separator)
print('Data Retrieval Complete')
print(separator)
```

## Creating Data Frame

```python
# creating a dictionary and data frame
weather_dict = {
    "city": name,
    "country": ctry,
    "lat": latitude,
    "temp": temperature,
    "humid": humidity,
    "cloud": cloudiness,
    "wind": windspeed
}
# creating the data frame
weather_data = pd.DataFrame(weather_dict)

# displaying the data frame
weather_data.head()
```

```python
# counting the values for each key/column
count = weather_data.count()
display(count)

# exporting the data frame to a csv file
weather_data.to_csv('weather_data.csv', sep=',', encoding='utf-8')
```
